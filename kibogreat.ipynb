# Frontalization and Detection Hybrid CAPG GAN AND MTCNN

```python
# Import Packages
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
import os
from torchvision import transforms, models
from PIL import Image

# === Attention Mechanisms ===

# Self-Attention Block
class SelfAttention(nn.Module):
    def __init__(self, in_channels):
        super(SelfAttention, self).__init__()
        self.query_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.key_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.value_conv = nn.Conv2d(in_channels, in_channels, kernel_size=1)
        self.gamma = nn.Parameter(torch.zeros(1))

    def forward(self, x):
        batch_size, channels, height, width = x.size()
        query = self.query_conv(x).view(batch_size, -1, height * width).permute(0, 2, 1)
        key = self.key_conv(x).view(batch_size, -1, height * width)
        value = self.value_conv(x).view(batch_size, -1, height * width)

        attention = torch.bmm(query, key)
        attention = F.softmax(attention, dim=-1)

        out = torch.bmm(value, attention.permute(0, 2, 1))
        out = out.view(batch_size, channels, height, width)

        out = self.gamma * out + x
        return out

# === CAPG-GAN Model with Self Attention ===
class CAPG_GAN(nn.Module):
    def __init__(self):
        super(CAPG_GAN, self).__init__()
        self.encoder = self.build_encoder()
        self.pyramid_module = self.build_pyramid()
        self.context_module = self.build_context()
        self.self_attention = SelfAttention(512)  # Apply self-attention
        self.decoder = self.build_decoder()

    def build_encoder(self):
        return nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1),
            nn.ReLU()
        )

    def build_pyramid(self):
        return nn.ModuleList([
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU()
        ])

    def build_context(self):
        return nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=1),
            nn.ReLU()
        )

    def build_decoder(self):
        return nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()  # Use Tanh for output to scale it between -1 and 1
        )

    def forward(self, x):
        encoded = self.encoder(x)
        for layer in self.pyramid_module:
            encoded = layer(encoded) + encoded  # Residual connection
        context_features = self.context_module(encoded)
        
        attention_out = self.self_attention(context_features)  # Apply self-attention
        
        output = self.decoder(attention_out)
        return output

# === Couple-Agent Discriminator ===
class CoupleAgentDiscriminator(nn.Module):
    def __init__(self):
        super(CoupleAgentDiscriminator, self).__init__()
        self.agent1 = self.build_discriminator()
        self.agent2 = self.build_discriminator()

    def build_discriminator(self):
        return nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(256, 1, kernel_size=4, stride=2, padding=1),
            nn.Sigmoid()
        )

    def forward(self, x, pose_embedding=None):
        output1 = self.agent1(x)
        if pose_embedding is not None:
            output2 = self.agent2(pose_embedding)
            return output1, output2
        return output1

# === VGG19 Feature Extractor ===
class VGG19FeatureExtractor(nn.Module):
    def __init__(self):
        super(VGG19FeatureExtractor, self).__init__()
        vgg19_model = models.vgg19(pretrained=True).features
        self.vgg19 = vgg19_model[:16]  # Use only the first few layers
        for param in self.vgg19.parameters():
            param.requires_grad = False  # Freeze VGG layers

    def forward(self, x):
        return self.vgg19(x)

# === Data Handling ===
class CustomDataset(Dataset):
    def __init__(self, off_angle_dir, frontal_dir, transform=None):
        self.off_angle_images = sorted([os.path.join(off_angle_dir, img) for img in os.listdir(off_angle_dir)])
        self.frontal_images = sorted([os.path.join(frontal_dir, img) for img in os.listdir(frontal_dir)])
        self.off_angle_images = self.off_angle_images[:len(self.frontal_images)]  # Match lengths
        self.transform = transform

    def __len__(self):
        return len(self.off_angle_images)

    def __getitem__(self, idx):
        off_angle_image = Image.open(self.off_angle_images[idx]).convert("RGB")
        frontal_image = Image.open(self.frontal_images[idx]).convert("RGB")
        if self.transform:
            off_angle_image = self.transform(off_angle_image)
            frontal_image = self.transform(frontal_image)
        return off_angle_image, frontal_image

# Data preprocessing with augmentations
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.Resize((128, 128)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

# Load dataset
dataset = CustomDataset("/kaggle/input/kibooo/CAS-PEAL-R1/pose", "/kaggle/input/kibooo/CAS-PEAL-R1/frontal", transform=transform)

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, val_size, test_size])

# Updated batch size from 32 to 64
batch_size = 64  # Increased batch size
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)

# === Loss Functions ===
criterion_GAN = nn.BCELoss()
criterion_content = nn.MSELoss()
criterion_L2 = nn.MSELoss()

# === Multi-scale Pixel-wise Loss ===
def multi_scale_pixel_loss(fake_images, real_images):
    scales = [1, 2, 4]  # Analyzing at multiple scales
    total_loss = 0
    for scale in scales:
        fake_scaled = F.interpolate(fake_images, scale_factor=1 / scale, mode='bilinear', align_corners=False)
        real_scaled = F.interpolate(real_images, scale_factor=1 / scale, mode='bilinear', align_corners=False)
        total_loss += criterion_L2(fake_scaled, real_scaled)
    return total_loss

# === Training Setup ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
generator = CAPG_GAN().to(device)
discriminator = CoupleAgentDiscriminator().to(device)
vgg19_extractor = VGG19FeatureExtractor().to(device)

# Adjusted learning rates
learning_rate_G = 1e-4
learning_rate_D = 1e-4

optimizer_G = torch.optim.Adam(generator.parameters(), lr=learning_rate_G, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=learning_rate_D, betas=(0.5, 0.999))

num_epochs = 150  # Increased number of epochs
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)

# === Save and Visualize Generated Images ===
def save_generated_images(fake_images, epoch, output_dir, num_images=5):
    fake_images = fake_images.cpu().detach()
    batch_size = fake_images.size(0)
    num_images = min(num_images, batch_size)
    for i in range(num_images):
        img = fake_images[i].permute(1, 2, 0).numpy()
        img = (img + 1) / 2  # Convert to range [0, 1]
        img = (img * 255).astype('uint8')  # Convert to uint8
        img_pil = Image.fromarray(img)
        img_pil.save(os.path.join(output_dir, f"image_epoch_{epoch}_{i}.png"))

# === Training Loop ===
for epoch in range(num_epochs):
    for i, (real_images, target_images) in enumerate(train_loader):
        real_images, target_images = real_images.to(device), target_images.to(device)

        # Train Discriminator
        discriminator.zero_grad()
        output_real = discriminator(target_images)
        loss_D_real = criterion_GAN(output_real, torch.ones_like(output_real))

        fake_images = generator(real_images)
        output_fake = discriminator(fake_images.detach())
        loss_D_fake = criterion_GAN(output_fake, torch.zeros_like(output_fake))

        loss_D = (loss_D_real + loss_D_fake) / 2
        loss_D.backward()
        optimizer_D.step()

        # Train Generator
        generator.zero_grad()
        output_fake_for_G = discriminator(fake_images)
        loss_G_GAN = criterion_GAN(output_fake_for_G, torch.ones_like(output_fake_for_G))

        real_features = vgg19_extractor(target_images)
        fake_features = vgg19_extractor(fake_images)
        loss_G_content = criterion_content(fake_features, real_features)

        loss_G_pixel = multi_scale_pixel_loss(fake_images, target_images)
        loss_G_L2 = criterion_L2(fake_images, target_images)

        # Adjusted weights for generator loss
        loss_G = loss_G_GAN + 0.05 * loss_G_content + 0.5 * loss_G_pixel + 0.1 * loss_G_L2
        loss_G.backward()
        optimizer_G.step()

        if i % 50 == 0:
            print(f"Epoch [{epoch}/{num_epochs}], Batch [{i}/{len(train_loader)}], Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}")

    # Save generated images at every epoch
    save_generated_images(fake_images, epoch, output_dir, num_images=5)

# Save final model
torch.save(generator.state_dict(), "generator_final.pth")
torch.save(discriminator.state_dict(), "discriminator_final.pth")

# Import Packages
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
import os
from torchvision import transforms, models
from PIL import Image

# === Attention Mechanisms ===

# Self-Attention Block
class SelfAttention(nn.Module):
    def __init__(self, in_channels):
        super(SelfAttention, self).__init__()
        self.query_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.key_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.value_conv = nn.Conv2d(in_channels, in_channels, kernel_size=1)
        self.gamma = nn.Parameter(torch.zeros(1))

    def forward(self, x):
        batch_size, channels, height, width = x.size()
        query = self.query_conv(x).view(batch_size, -1, height * width).permute(0, 2, 1)
        key = self.key_conv(x).view(batch_size, -1, height * width)
        value = self.value_conv(x).view(batch_size, -1, height * width)

        attention = torch.bmm(query, key)
        attention = F.softmax(attention, dim=-1)

        out = torch.bmm(value, attention.permute(0, 2, 1))
        out = out.view(batch_size, channels, height, width)

        out = self.gamma * out + x
        return out

# === CAPG-GAN Model with Self Attention ===
class CAPG_GAN(nn.Module):
    def __init__(self):
        super(CAPG_GAN, self).__init__()
        self.encoder = self.build_encoder()
        self.pyramid_module = self.build_pyramid()
        self.context_module = self.build_context()
        self.self_attention = SelfAttention(512)  # Apply self-attention
        self.decoder = self.build_decoder()

    def build_encoder(self):
        return nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1),
            nn.ReLU()
        )

    def build_pyramid(self):
        return nn.ModuleList([
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU()
        ])

    def build_context(self):
        return nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=1),
            nn.ReLU()
        )

    def build_decoder(self):
        return nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()  # Use Tanh for output to scale it between -1 and 1
        )

    def forward(self, x):
        encoded = self.encoder(x)
        for layer in self.pyramid_module:
            encoded = layer(encoded) + encoded  # Residual connection
        context_features = self.context_module(encoded)
        
        # Apply Self-Attention only
        attention_out = self.self_attention(context_features)  # Apply self-attention
        
        output = self.decoder(attention_out)
        return output

# === Couple-Agent Discriminator ===
class CoupleAgentDiscriminator(nn.Module):
    def __init__(self):
        super(CoupleAgentDiscriminator, self).__init__()
        self.agent1 = self.build_discriminator()
        self.agent2 = self.build_discriminator()

    def build_discriminator(self):
        return nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(256, 1, kernel_size=4, stride=2, padding=1),
            nn.Sigmoid()
        )

    def forward(self, x, pose_embedding=None):
        output1 = self.agent1(x)
        if pose_embedding is not None:
            output2 = self.agent2(pose_embedding)
            return output1, output2
        return output1

# === VGG19 Feature Extractor ===
class VGG19FeatureExtractor(nn.Module):
    def __init__(self):
        super(VGG19FeatureExtractor, self).__init__()
        vgg19_model = models.vgg19(pretrained=True).features
        self.vgg19 = vgg19_model[:16]  # Use only the first few layers
        for param in self.vgg19.parameters():
            param.requires_grad = False  # Freeze VGG layers

    def forward(self, x):
        return self.vgg19(x)

# === Data Handling ===
class CustomDataset(Dataset):
    def __init__(self, off_angle_dir, frontal_dir, transform=None):
        self.off_angle_images = sorted([os.path.join(off_angle_dir, img) for img in os.listdir(off_angle_dir)])
        self.frontal_images = sorted([os.path.join(frontal_dir, img) for img in os.listdir(frontal_dir)])
        self.off_angle_images = self.off_angle_images[:len(self.frontal_images)]  # Match lengths
        self.transform = transform

    def __len__(self):
        return len(self.off_angle_images)

    def __getitem__(self, idx):
        off_angle_image = Image.open(self.off_angle_images[idx]).convert("RGB")
        frontal_image = Image.open(self.frontal_images[idx]).convert("RGB")
        if self.transform:
            off_angle_image = self.transform(off_angle_image)
            frontal_image = self.transform(frontal_image)
        return off_angle_image, frontal_image

# Data preprocessing
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

# Load dataset
dataset = CustomDataset("/kaggle/input/kibooo/CAS-PEAL-R1/pose",
                        "/kaggle/input/kibooo/CAS-PEAL-R1/frontal", transform=transform)

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, val_size, test_size])
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
test_loader = DataLoader(test_dataset, batch_size=32)

# === Loss Functions ===
criterion_GAN = nn.BCELoss()
criterion_content = nn.MSELoss()
criterion_L2 = nn.MSELoss()

# === Multi-scale Pixel-wise Loss ===
def multi_scale_pixel_loss(fake_images, real_images):
    scales = [1, 2, 4]
    total_loss = 0
    for scale in scales:
        fake_scaled = F.interpolate(fake_images, scale_factor=1 / scale, mode='bilinear', align_corners=False)
        real_scaled = F.interpolate(real_images, scale_factor=1 / scale, mode='bilinear', align_corners=False)
        total_loss += criterion_L2(fake_scaled, real_scaled)
    return total_loss

# === Training Setup ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
generator = CAPG_GAN().to(device)
discriminator = CoupleAgentDiscriminator().to(device)
vgg19_extractor = VGG19FeatureExtractor().to(device)
optimizer_G = torch.optim.Adam(generator.parameters(), lr=2e-4, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=2e-4, betas=(0.5, 0.999))

num_epochs = 100
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)

# === Save and Visualize Generated Images ===
def save_generated_images(fake_images, epoch, output_dir, num_images=5):
    fake_images = fake_images.cpu().detach()
    batch_size = fake_images.size(0)
    num_images = min(num_images, batch_size)
    for i in range(num_images):
        img = fake_images[i].permute(1, 2, 0).numpy()
        img = (img + 1) / 2  # Convert to range [0, 1]
        img = (img * 255).astype('uint8')  # Convert to uint8
        img_pil = Image.fromarray(img)
        img_pil.save(os.path.join(output_dir, f"image_epoch_{epoch}_{i}.png"))

# === Training Loop ===
for epoch in range(num_epochs):
    for i, (real_images, target_images) in enumerate(train_loader):
        real_images, target_images = real_images.to(device), target_images.to(device)

        # Train Discriminator
        discriminator.zero_grad()
        output_real = discriminator(target_images)
        loss_D_real = criterion_GAN(output_real, torch.ones_like(output_real))

        fake_images = generator(real_images)
        output_fake = discriminator(fake_images.detach())
        loss_D_fake = criterion_GAN(output_fake, torch.zeros_like(output_fake))

        loss_D = (loss_D_real + loss_D_fake) / 2
        loss_D.backward()
        optimizer_D.step()

        # Train Generator
        generator.zero_grad()
        output_fake_for_G = discriminator(fake_images)
        loss_G_GAN = criterion_GAN(output_fake_for_G, torch.ones_like(output_fake_for_G))

        real_features = vgg19_extractor(target_images)
        fake_features = vgg19_extractor(fake_images)
        loss_G_content = criterion_content(fake_features, real_features)

        loss_G_pixel = multi_scale_pixel_loss(fake_images, target_images)
        loss_G_L2 = criterion_L2(fake_images, target_images)

        loss_G = loss_G_GAN + 0.1 * loss_G_content + 0.5 * loss_G_pixel + 0.1 * loss_G_L2
        loss_G.backward()
        optimizer_G.step()

        if i % 50 == 0:
            print(f"Epoch [{epoch}/{num_epochs}], Batch [{i}/{len(train_loader)}], Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}")

    save_generated_images(fake_images, epoch, output_dir, num_images=5)

# Save final model
torch.save(generator.state_dict(), "generator_final.pth")
torch.save(discriminator.state_dict(), "discriminator_final.pth")
```

```python
import os
import re
import numpy as np
import cv2
import torch
import matplotlib.pyplot as plt
from PIL import Image
from facenet_pytorch import MTCNN
from math import ceil

# Initialize MTCNN model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(keep_all=True, device=device)

# Directory containing generated images (Kaggle Input)
image_dir = "/kaggle/working/generated_images"  # Directory for input images
output_dir_train = "/kaggle/working/detected_faces/train"  # Output directory for training images
output_dir_test = "/kaggle/working/detected_faces/test"  # Output directory for testing images
output_dir_train_faces = "/kaggle/working/detected_faces/train_faces"  # Directory for training face images
output_dir_test_faces = "/kaggle/working/detected_faces/test_faces"  # Directory for testing face images

# Ensure the output directories exist
os.makedirs(output_dir_train, exist_ok=True)
os.makedirs(output_dir_test, exist_ok=True)
os.makedirs(output_dir_train_faces, exist_ok=True)
os.makedirs(output_dir_test_faces, exist_ok=True)

# Function to extract epoch number from filename
def extract_epoch(filename):
    match = re.search(r'(\d+)', filename)  # Extracts numbers from filename
    return int(match.group(1)) if match else -1  # Default to -1 if no number found

# Function to detect faces and landmarks
def detect_faces_and_landmarks(img):
    boxes, probs, landmarks = mtcnn.detect(img, landmarks=True)
    return boxes, landmarks

# Function to visualize detected faces, save images with bounding boxes, landmarks, and localized faces
def process_and_save_images(image_files, output_directory, output_faces_directory):
    for filename in image_files:
        img_path = os.path.join(image_dir, filename)
        img = Image.open(img_path).convert("RGB")
        img_np = np.array(img)

        # Detect faces and landmarks
        boxes, landmarks = detect_faces_and_landmarks(img)

        # Draw bounding boxes and landmarks on the image
        if boxes is not None:
            for j, box in enumerate(boxes):
                x1, y1, x2, y2 = map(int, box)

                # Define adjustments to exclude ears and forehead
                ear_margin = 7  # Margin to exclude ears
                forehead_margin = 15  # Margin to exclude forehead

                # Adjust boundaries to not include ears or forehead
                x1 = max(0, x1 + ear_margin)  # Move right for ears
                x2 = min(img_np.shape[1] - 1, x2 - ear_margin)  # Move left for ears
                y1 = min(y1 + forehead_margin, y2)  # Move down for forehead

                # Validate that the box still has valid dimensions
                if x2 <= x1 or y2 <= y1:
                    continue  # Skip if we get an invalid box

                # Draw bounding box
                cv2.rectangle(img_np, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box

                # Draw landmarks
                if landmarks is not None and isinstance(landmarks, np.ndarray) and landmarks.ndim == 3:
                    for point in landmarks[j]:
                        cv2.circle(img_np, tuple(point.astype(int)), 3, (255, 0, 0), -1)  # Red points

                # Save the localized face
                face_img = img_np[y1:y2, x1:x2]  # Extract the face region
                if face_img.size > 0:  # Check if the face image is valid
                    face_output_path = os.path.join(output_faces_directory, f"face_{filename[:-4]}_{j}.jpg")
                    cv2.imwrite(face_output_path, cv2.cvtColor(face_img, cv2.COLOR_RGB2BGR))

        # Save the processed image with bounding boxes and landmarks
        img_np_bgr = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)
        output_path = os.path.join(output_directory, f"detected_{filename}")
        cv2.imwrite(output_path, img_np_bgr)

# Load all image files
image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Sort images by epoch number (ascending order)
image_files.sort(key=extract_epoch)

# Split the images into training and testing sets (70% training, 30% testing)
split_index = int(0.7 * len(image_files))
train_images = image_files[:split_index]
test_images = image_files[split_index:]

# Process and save training and testing images
process_and_save_images(train_images, output_dir_train, output_dir_train_faces)
process_and_save_images(test_images, output_dir_test, output_dir_test_faces)

# If you want to display only the last 24 images from the training set, we can visualize them after processing
last_24_images = train_images[-24:]

# Visualize the last 24 processed images
def visualize_last_images(image_files, output_directory):
    num_images = len(image_files)
    cols = 4
    rows = ceil(num_images / cols)

    plt.figure(figsize=(cols * 4, rows * 4))

    for i in range(num_images):
        output_path = os.path.join(output_directory, f"detected_{image_files[i]}")
        img_np = cv2.imread(output_path)
        img_np = cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB)

        ax = plt.subplot(rows, cols, i + 1)
        ax.imshow(img_np)
        ax.axis("off")
        ax.set_title(f"Epoch {extract_epoch(image_files[i])}", fontsize=10)

    plt.tight_layout()
    plt.show()

# Visualize the last 24 processed images from the training set
visualize_last_images(last_24_images, output_dir_train)

print(f"Processed training images saved in: {output_dir_train}")
print(f"Processed testing images saved in: {output_dir_test}")
print(f"Localized faces saved in: {output_dir_train_faces} (train) and {output_dir_test_faces} (test)")
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pyplot as plt

# Face Detection Libraries
from mtcnn import MTCNN
from retinaface import RetinaFace

# Paths
pose_path = '/kaggle/input/sussess/multipie/pose'

# Map camera to yaw
camera_to_angle = {
    '8': 15, '9': 30, '10': 45, '11': 60, '12': 75, '13': 90,
    '7': 0  # frontal (for ground truth)
}

# Initialize detectors
mtcnn = MTCNN()
haar = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")  # proxy for DDFD

def get_bbox_mtcnn(image):
    dets = mtcnn.detect_faces(np.array(image))
    return dets[0]['box'] if dets else None

def get_bbox_ddfd(image):
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)
    faces = haar.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
    return faces[0] if len(faces) > 0 else None

def get_bbox_retinaface(image):
    dets = RetinaFace.detect_faces(np.array(image))
    if isinstance(dets, dict) and dets:
        key = list(dets.keys())[0]
        x1, y1, x2, y2 = map(int, dets[key]['facial_area'])
        return [x1, y1, x2 - x1, y2 - y1]
    return None

def bbox_to_coords(bbox):
    if bbox is None:
        return None
    x, y, w, h = bbox
    return [x, y, x + w, y + h]

def compute_iou(boxA, boxB):
    if boxA is None or boxB is None:
        return 0.0
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    return interArea / float(boxAArea + boxBArea - interArea + 1e-5)

# Collect all files
files = [f for f in os.listdir(pose_path) if f.endswith('.jpg') or f.endswith('.png')]

# Step 1: Get frontal (0Â°) as ground truth
frontal_gt = {}
for file in files:
    parts = file.split('_')
    if len(parts) >= 4 and parts[3] == '7':  # camera 7 = 0Â°
        key = '_'.join(parts[:3])
        image = Image.open(os.path.join(pose_path, file))
        bbox = bbox_to_coords(get_bbox_mtcnn(image))
        frontal_gt[key] = bbox

# Step 2: Loop through pose images and evaluate
results = []

for file in files:
    parts = file.split('_')
    if len(parts) < 4:
        continue
    cam_id = parts[3]
    angle = camera_to_angle.get(cam_id)
    if angle is None or angle == 0:
        continue
    key = '_'.join(parts[:3])
    if key not in frontal_gt:
        continue

    gt = frontal_gt[key]
    img = Image.open(os.path.join(pose_path, file))

    for model_name, detector_fn in {
        'MTCNN': get_bbox_mtcnn,
        'DDFD': get_bbox_ddfd,
        'RetinaFace': get_bbox_retinaface
    }.items():
        try:
            pred = detector_fn(img)
            pred = bbox_to_coords(pred)
            iou = compute_iou(gt, pred)
            recall = 1 if iou >= 0.5 else 0
            results.append({'pose': angle, 'model': model_name, 'iou': iou, 'recall': recall})
        except Exception as e:
            print(f"Error in {model_name} on {file}: {e}")

# Step 3: Table
results_df = pd.DataFrame(results)
summary = results_df.groupby(['model', 'pose']).agg({'iou': 'mean', 'recall': 'mean'}).reset_index()
pivot_iou = summary.pivot(index='model', columns='pose', values='iou')
pivot_recall = summary.pivot(index='model', columns='pose', values='recall')

# Combine IoU and Recall in one table
final_table = []
for model in pivot_iou.index:
    final_table.append([model, 'IOU'] + [round(pivot_iou.loc[model, angle], 2) if angle in pivot_iou.columns else None for angle in [15, 30, 45, 60, 75, 90]])
    final_table.append(["", 'Recall'] + [round(pivot_recall.loc[model, angle], 2) if angle in pivot_recall.columns else None for angle in [15, 30, 45, 60, 75, 90]])

final_df = pd.DataFrame(final_table, columns=['Model', 'Evaluation Metrics', '15Â°', '30Â°', '45Â°', '60Â°', '75Â°', '90Â°'])
print(final_df)

# Optional: save to CSV
# final_df.to_csv("/kaggle/working/detection_metrics_summary.csv", index=False)
```

```python
import os
import numpy as np
import cv2
import torch
from PIL import Image
from facenet_pytorch import MTCNN

# Initialize devices and models
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(keep_all=True, device=device)
mtcnn_plus = MTCNN(keep_all=True, device=device, min_face_size=40, post_process=True)

# Define input and output directories
input_dirs = [
    "/kaggle/input/kookeet/multipie/pose"
]
output_dirs = [
    "/kaggle/working/detected_2"
]

# Ensure output directories exist
for output_dir in output_dirs:
    os.makedirs(output_dir, exist_ok=True)

# Function to detect faces and landmarks
def detect_faces_and_landmarks(img, model):
    boxes, _, landmarks = model.detect(img, landmarks=True)
    return boxes, landmarks

# Function to process images, detect faces, and save results
def process_and_save_images(image_dir, output_dir, model):
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    for filename in image_files:
        img_path = os.path.join(image_dir, filename)
        img = Image.open(img_path).convert("RGB")
        img_np = np.array(img)

        # Detect faces and landmarks
        boxes, landmarks = detect_faces_and_landmarks(img, model)

        if boxes is not None:
            for j, box in enumerate(boxes):
                x1, y1, x2, y2 = map(int, box)

                # Adjust boundaries to not include ears or forehead
                ear_margin = 7
                forehead_margin = 15
                x1 = max(0, x1 + ear_margin)
                x2 = min(img_np.shape[1] - 1, x2 - ear_margin)
                y1 = min(y1 + forehead_margin, y2)

                if x2 <= x1 or y2 <= y1:
                    continue

                # Draw bounding box
                cv2.rectangle(img_np, (x1, y1), (x2, y2), (0, 255, 0), 2)

                # Draw landmarks if detected
                if landmarks is not None:
                    for point in landmarks[j]:
                        cv2.circle(img_np, tuple(point.astype(int)), 3, (255, 0, 0), -1)

                # Save localized face with lower quality
                face_img = img_np[y1:y2, x1:x2]
                if face_img.size > 0:
                    face_output_path = os.path.join(output_dir, f"face_{filename[:-4]}_{j}.jpg")
                    # Save with lower quality (e.g., 30)
                    cv2.imwrite(face_output_path, cv2.cvtColor(face_img, cv2.COLOR_RGB2BGR), [int(cv2.IMWRITE_JPEG_QUALITY), 30])

            # Save the processed image with bounding boxes
            img_np_bgr = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)
            output_path = os.path.join(output_dir, f"detected_{filename}")
            cv2.imwrite(output_path, img_np_bgr)

# Process images using MTCNN and save to specified output directories
for input_dir, output_dir in zip(input_dirs, output_dirs):
    process_and_save_images(input_dir, output_dir, mtcnn)

# Optionally, also process images using MTCNN++ 
for input_dir, output_dir in zip(input_dirs, output_dirs):
    process_and_save_images(input_dir, output_dir, mtcnn_plus)

print("Processed images saved in the specified output directories in /kaggle/working/.")
```

```python
import os
import shutil
import zipfile

# Source and target directories
source_dir = '/kaggle/input/sussess/multipie'
target_dir = '/kaggle/working/multipie_sorted'

# Mapping: camera ID â†’ yaw pose
cam_pose_map = {
    '050': '0',
    '051': '15',
    '041': '30',
    '080': '60',
    '090': '60',
    '130': '75',
    '140': '90'
}

# Create target folders
for pose_label in sorted(set(cam_pose_map.values()), key=int):
    os.makedirs(os.path.join(target_dir, pose_label), exist_ok=True)

# Sort images
count = 0
for root, _, files in os.walk(source_dir):
    for file in files:
        for cam_id, pose_label in cam_pose_map.items():
            if f'cam{cam_id}' in file:
                src_path = os.path.join(root, file)
                dest_path = os.path.join(target_dir, pose_label, file)
                shutil.copy(src_path, dest_path)
                count += 1
                print(f"âœ… {file} â†’ Pose_{pose_label}")
                break

print(f"\nâœ… Done: {count} images sorted by pose into {target_dir}")

# Zip the sorted directory
zip_path = '/kaggle/working/multipie_sorted.zip'
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, _, files in os.walk(target_dir):
        for file in files:
            abs_path = os.path.join(root, file)
            rel_path = os.path.relpath(abs_path, start=target_dir)
            zipf.write(abs_path, rel_path)

print(f"\nðŸŽ‰ Zip created at: {zip_path}")
```

```python
import os
from pathlib import Path
import shutil
import zipfile

# Input and output paths
input_dir = Path("/kaggle/input/sussess/multipie/pose")
output_base = Path("/kaggle/working/sorted_multipie")

# Pose ID â†’ Yaw angle mapping
pose_id_to_yaw = {
    "051": 0,
    "060": 15,
    "041": 30,
    "050": 45,
    "042": 60,
    "070": 75,
    "082": 90
}

# Create output folders for each yaw angle
for yaw in sorted(set(pose_id_to_yaw.values())):
    (output_base / f"{yaw}").mkdir(parents=True, exist_ok=True)

# Copy images to appropriate yaw folders
for file in input_dir.glob("*.jpg"):
    parts = file.stem.split("_")
    if len(parts) < 5:
        continue
    pose_id = parts[3]
    yaw = pose_id_to_yaw.get(pose_id)
    if yaw is not None:
        dest = output_base / f"{yaw}" / file.name
        shutil.copy(file, dest)

# Zip the sorted folder
zip_path = "/kaggle/working/sorted_multipie.zip"
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for folder_path, _, filenames in os.walk(output_base):
        for filename in filenames:
            file_path = os.path.join(folder_path, filename)
            arcname = os.path.relpath(file_path, output_base)
            zipf.write(file_path, arcname)

print("âœ… Sorting and zipping complete. Saved to /kaggle/working/sorted_multipie.zip")
```

```python
import os
import shutil
import zipfile

def parse_pose_from_filename(filename):
    parts = filename.split('_')
    if len(parts) != 6:
        return None
    try:
        pose_num = int(parts[3])
    except ValueError:
        return None
    return pose_num

def closest_pose(pose, target_poses):
    return min(target_poses, key=lambda x: abs(x - pose))

def read_sort_and_save(input_folder, output_folder, target_poses, zip_path):
    # List all jpg files in input folder
    files = [f for f in os.listdir(input_folder) if f.lower().endswith('.jpg')]
    matched_files = []

    for f in files:
        pose_num = parse_pose_from_filename(f)
        if pose_num is None:
            print(f"Warning: Could not parse pose from filename: {f}")
            continue
        matched = closest_pose(pose_num, target_poses)
        matched_files.append({'filename': f, 'matched_pose': matched})

    # Sort by matched pose
    matched_files.sort(key=lambda x: x['matched_pose'])

    # Create output folders for each target pose
    for pose in target_poses:
        folder_path = os.path.join(output_folder, str(pose).zfill(3))
        os.makedirs(folder_path, exist_ok=True)

    # Copy files into corresponding pose folder
    for item in matched_files:
        src = os.path.join(input_folder, item['filename'])
        dst_folder = os.path.join(output_folder, str(item['matched_pose']).zfill(3))
        dst = os.path.join(dst_folder, item['filename'])
        shutil.copy2(src, dst)

    # Create ZIP archive of output folder
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(output_folder):
            for file in files:
                abs_path = os.path.join(root, file)
                rel_path = os.path.relpath(abs_path, output_folder)
                zipf.write(abs_path, rel_path)

    print(f"Processed {len(matched_files)} images.")
    print(f"Images saved to: {output_folder}")
    print(f"ZIP archive created at: {zip_path}")


# === PARAMETERS ===
input_folder = '/kaggle/input/sussess/multipie/pose/'
output_folder = '/kaggle/working/multipie_matched_pose/'
target_poses = [0, 15, 30, 45, 60, 90]
zip_path = '/kaggle/working/multipie_matched_pose.zip'

# === RUN SCRIPT ===
read_sort_and_save(input_folder, output_folder, target_poses, zip_path)
```
